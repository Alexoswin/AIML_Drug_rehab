```python
#label encoding

from sklearn.preprocessing  import LabelEncoder
label_encoder = LabelEncoder()
columns_to_encode=['Gender', 'Educated', 'Living with', 'Conflict with Law',
       'Criminal Record', 'Spends Most Time With', 'Failure in Life',
       'Mental/Emotional Problem', 'Suicidal Thoughts', 'Family Relation',
       'Financial Status', 'Addicted Persons in Family','Satisfied with Workspace', 'Living with Drug Consumer', 'Smoking','City',
       'Influenced By', 'Ease to Control Drug Use',
       'Type of Drug Consumed']

def encode(df, columns_to_encode):
  for index in columns_to_encode:
      df[index] = label_encoder.fit_transform(df[index])

encode(df,columns_to_encode)
df.head()
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split

features= df[['Age', 'Gender', 'Educated', 'Living with', 'Conflict with Law',
       'Criminal Record', 'Spends Most Time With', 'Failure in Life',
       'Mental/Emotional Problem', 'Suicidal Thoughts', 'Family Relation',
       'Financial Status', 'Addicted Persons in Family', 'No. of Friends',
       'Satisfied with Workspace', 'Living with Drug Consumer', 'Smoking',
       'Influenced By', 'Ease to Control Drug Use',
       'Frequency of Drug Usage per Week',  'City',
       'Type of Drug Consumed']]

target =df['Days Taken to Get Cure']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


preprocessor = ColumnTransformer(
    transformers=[
        # No encoder specified for categorical variables
        ('passthrough', 'passthrough', ['Gender', 'Educated', 'Living with',
                                        'Conflict with Law', 'Criminal Record',
                                        'Spends Most Time With', 'Failure in Life',
                                        'Mental/Emotional Problem', 'Suicidal Thoughts',
                                        'Family Relation', 'Financial Status',
                                        'Addicted Persons in Family', 'Satisfied with Workspace',
                                        'Living with Drug Consumer', 'Smoking', 'Influenced By',
                                        'Ease to Control Drug Use', 'City',
                                        'Type of Drug Consumed'])
    ],
    remainder='passthrough'
)

# Define the Decision Tree model pipeline
tree_model = Pipeline([
    ('preprocessor', preprocessor),
    ('model', DecisionTreeRegressor())  # Using DecisionTreeRegressor for regression task
])

# Train the model
tree_model.fit(X_train, y_train)

# Make predictions
tree_predictions = tree_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print("Decision Tree Accuracy:", accuracy_score(y_test, tree_predictions))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
forest_model = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier())
])

forest_model.fit(X_train, y_train)
forest_predictions = forest_model.predict(X_test)
```

```python
print("Random forest accuracy:", accuracy_score(y_test, forest_predictions))
```

```python
# Support Vector Machine (SVM)
from sklearn.svm import SVC
svm_model = Pipeline([
    ('preprocessor', preprocessor),
    ('model', SVC())
])

svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
```

```python
print("support vector:", accuracy_score(y_test, svm_predictions))
```

```python
# k-Nearest Neighbors (k-NN)
from sklearn.neighbors import KNeighborsClassifier
knn_model = Pipeline([
    ('preprocessor', preprocessor),
    ('model', KNeighborsClassifier())
])

knn_model.fit(X_train, y_train)
knn_predictions = knn_model.predict(X_test)
```

```python
print("support vector:", accuracy_score(y_test, knn_predictions))
```

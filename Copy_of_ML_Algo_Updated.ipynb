```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("/content/drugAddiction (1).csv")
df.head(100)
```

```python
df = df.dropna()

df = df.drop(df[df["Days Taken to Get Cure"]<1].index)

df['Days Taken to Get Cure'] = (df['Days Taken to Get Cure']/7).astype(int)
```

```python
df.dtypes
```

```python
print(df.columns)
```

```python
#label encoding

from sklearn.preprocessing  import LabelEncoder
label_encoder = LabelEncoder()
columns_to_encode=['Age', 'Gender', 'Education', 'Live with', 'Spend most time',
       'Failure in life', 'Mental/emotional problem', 'Suicidal thoughts',
       'Family relationship', 'Financials of family', 'Influenced By',
       'Satisfied with workplace', 'Easy to control use of drug',
         'Type of Drug Consumed']

def encode(df, columns_to_encode):
  for index in columns_to_encode:
      df[index] = label_encoder.fit_transform(df[index])

encode(df,columns_to_encode)
df.head()
```

```python
df.head(100)
```

```python
df.isna().any().any()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split

features= df[['Age', 'Gender', 'Education', 'Live with', 'Spend most time',
       'Failure in life', 'Mental/emotional problem', 'Suicidal thoughts',
       'Family relationship', 'Financials of family', 'Influenced By',
       'Satisfied with workplace', 'Easy to control use of drug',
       'Days Taken to Get Cure', 'Frequency of Drug Usage per Week',
       'Type of Drug Consumed']]

target =df['Days Taken to Get Cure']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


preprocessor = ColumnTransformer(
    transformers=[
        # No encoder specified for categorical variables
        ('passthrough', 'passthrough', ['Age', 'Gender', 'Education', 'Live with', 'Spend most time',
       'Failure in life', 'Mental/emotional problem', 'Suicidal thoughts',
       'Family relationship', 'Financials of family', 'Influenced By',
       'Satisfied with workplace', 'Easy to control use of drug',
       'Days Taken to Get Cure', 'Frequency of Drug Usage per Week',
       'Type of Drug Consumed'])
    ],
    remainder='passthrough'
)

```

***KNN***

```python
# k-Nearest Neighbors (k-NN)
from sklearn.neighbors import KNeighborsClassifier
knn_model = Pipeline([
    ('preprocessor', preprocessor),
    ('model', KNeighborsClassifier())
])

knn_model.fit(X_train, y_train)
knn_predictions = knn_model.predict(X_test)
```

```python

from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming you have defined your KNN model with appropriate parameters
knn_model = KNeighborsRegressor()

# Define your pipeline with the KNN model
pipeline = Pipeline([('knn', knn_model)])

# Define the number of folds
k = 5

# Initialize the KFold splitter
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Perform cross-validation
mse_scores = cross_val_score(pipeline, features, target, cv=kf, scoring='neg_mean_squared_error')

# Convert negative MSE to positive and calculate RMSE
rmse_scores = np.sqrt(-mse_scores)

# Print the cross-validation RMSE scores
print("Cross-Validation RMSE Scores:", rmse_scores)

# Calculate and print the mean and standard deviation of the RMSE scores
print("Mean RMSE:", np.mean(rmse_scores))
print("Standard Deviation of RMSE:", np.std(rmse_scores))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Fit the KNN model to the training data
knn_model.fit(X_train, y_train)

# Compute predictions
predictions = knn_model.predict(X_test)

# Compute evaluation metrics

# We can use regression evaluation metrics instead.
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print()
# Compute evaluation metrics


```

***DECISION*** ***TREE***

```python
# Define the Decision Tree model pipeline
tree_model = Pipeline([
    ('preprocessor', preprocessor),
    ('model', DecisionTreeRegressor())  # Using DecisionTreeRegressor for regression task
])

# Train the model
tree_model.fit(X_train, y_train)

# Make predictions
tree_predictions = tree_model.predict(X_test)
```

```python

from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
import numpy as np

# Instantiate the Decision Tree model
model = DecisionTreeRegressor(random_state=42)

# Define the number of folds
k = 5

# Initialize the KFold splitter
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Perform cross-validation
mse_scores = cross_val_score(model, features, target, cv=kf, scoring='neg_mean_squared_error')

# Convert negative MSE to positive and calculate RMSE
rmse_scores = np.sqrt(-mse_scores)

# Print the cross-validation RMSE scores
print("Cross-Validation RMSE Scores:", rmse_scores)

# Calculate and print the mean and standard deviation of the RMSE scores
print("Mean RMSE:", np.mean(rmse_scores))
print("Standard Deviation of RMSE:", np.std(rmse_scores))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Compute predictions
predictions = tree_model.predict(X_test)

# Compute evaluation metrics
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')
f1 = f1_score(y_test, predictions, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_absolute_error
import numpy as np
from sklearn.ensemble import RandomForestClassifier

# Instantiate the RandomForestClassifier model
model = RandomForestClassifier(random_state=42)

# Define the number of folds
k = 5

# Initialize the KFold splitter
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Perform cross-validation with Mean Absolute Error (MAE)
mae_scores = cross_val_score(model, features, target, cv=kf, scoring='neg_mean_absolute_error')

# Convert negative MAE to positive and calculate Mean Absolute Error
mae_scores = -mae_scores

# Print the cross-validation MAE scores
print("Cross-Validation MAE Scores:", mae_scores)

# Calculate and print the mean and standard deviation of the MAE scores
print("Mean MAE:", np.mean(mae_scores))
print("Standard Deviation of MAE:", np.std(mae_scores))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.ensemble import RandomForestClassifier

# Create and fit the RandomForestClassifier model
forest_model = RandomForestClassifier()
forest_model.fit(X_train, y_train)

# Compute predictions
forest_predictions = forest_model.predict(X_test)

# Compute evaluation metrics
accuracy = accuracy_score(y_test, forest_predictions)
precision = precision_score(y_test, forest_predictions, average='weighted')
recall = recall_score(y_test, forest_predictions, average='weighted')
f1 = f1_score(y_test, forest_predictions, average='weighted')

# Print evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
